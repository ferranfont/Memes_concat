```python
import pandas as pd

# Especifica el nombre del archivo
filename = '0xe436F5a05D0D132058517B7f68f07c05B88d8820.csv'

# Define el directorio donde se encuentra el archivo
directory = 'processed_csv/'

# Construye la ruta cFompleta al archivo
file_path = directory + filename

# Lee el archivo CSV en un DataFrame
try:
    df = pd.read_csv(file_path, skiprows=1, encoding='utf-8')
    print("Archivo convertido a DataFrame con éxito.")
except FileNotFoundError:
    print(f"No se encontró el archivo: {file_path}")
except Exception as e:
    print(f"Ocurrió un error al leer el archivo: {e}")

filename_without_extension = filename.split('.')[0]
df['Filename'] = filename_without_extension
```
